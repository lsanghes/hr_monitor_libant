```python
import matplotlib.pyplot as plt, pandas as pd, csv
file = "hr_log.csv_20171211"
hr = pd.read_csv(file, header=None, names=["time", "hr","avg"])
hr["ts"] = pd.to_datetime(hr["time"], format="%m/%d/%Y %H:%M:%S")
```

```python
plt.figure(figsize=(15,4))
plt.plot(hr['ts'], hr['hr'], '-')
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.plot(hr['ts'], hr['avg'], '-')
plt.show()
```
